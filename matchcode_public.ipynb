```python
import gurobipy as grb
import pandas as pd
import numpy as np
mp=pd.read_excel('match data.xlsx') 
pd.set_option("display.max_columns",100)  
```

```python
mp=mp.reset_index()
```

```python
# set dictionary
P={}

# define lists of choice options
Lo=["North West Los Angeles (Burbank, San Fernando)",
"West (Santa Monica, Venice, Westwood)",
"Central  (Downtown LA, Koreantown, around USC)",
"South  (San Pedro, Long Beach)",
"East (San Gabriel Valley, Pasadena)",
"San Diego",
"Santa Barbara",
"North Orange County",
"South Orange County",
"Ventura"]
Ti=["Morning/breakfast",
"Lunch",
"Afternoon/Early Evening/Happy Hour",
"Coffee Hour (Late Morning/Afternoon)",
"Week days",
"Weekends",
"Virtual (phone, video, etc.)",
"Before/After WiBD Meetups, events"]
ForMentee = ["Technical skills to polish",
"How to break into the field of big data/analytics",
"Career prep such as job/intern searching, job advancement, resume building, networking",
"Managing a career and family",
"Industry insights and current challenges",
"Newest technology trend or technical tools used in the industry"]
FromMentor = ["Technical skills to polish",
"How to break into the field of big data/analytics",
"Career prep such as job/intern searching, job advancement, resume building, networking",
"Managing a career and family",
"Industry insights and current challenges",
"Newest technology trend or technical tools used in the industry"]
FuncEx=["Accounting",
"Analytics",
"Consulting (external)",
"Consulting (internal)",
"Consumer Products",
"Education",
"Engineering",
"Finance",
"Human Capital/Workforce",
"Information Technology",
"Marketing",
"Operations/Supply Chain",
"Product Management",
"Sales"]
FuncIn=["Accounting",
"Analytics",
"Consulting (external)",
"Consulting (internal)",
"Consumer Products",
"Education",
"Engineering",
"Finance",
"Human Capital/Workforce",
"Information Technology",
"Marketing",
"Operations/Supply Chain",
"Product Management",
"Sales"]
IndEx = ["Aerospace/Defense",
"Agriculture",
"Education",
"Energy",
"Entertainment",
"Fashion",
"Financial Services",
"Gaming",
"Government",
"Healthcare",
"Insurance",
"Internet",
"Legal Services",
"Manufacturing",
"Non-Profit",
"Pharma/Bio-Tech",
"Politics",
"Retail/Wholesale",
"Social Services",
"Technology/Software",
"Travel/Tourism/Hospitality",
"Transportation & Logistics"]
IndIn = ["Aerospace/Defense",
"Agriculture",
"Education",
"Energy",
"Entertainment",
"Fashion",
"Financial Services",
"Gaming",
"Government",
"Healthcare",
"Insurance",
"Internet",
"Legal Services",
"Manufacturing",
"Non-Profit",
"Pharma/Bio-Tech",
"Politics",
"Retail/Wholesale",
"Social Services",
"Technology/Software",
"Travel/Tourism/Hospitality",
"Transportation & Logistics"]
AreaEx = ["Analytical tools and software",
"Business Analytics",
"Data Science",
"Machine Learning, Artificial Intelligence",
"Software Engineering"]
AreaIn = ["Analytical tools and software",
"Business Analytics",
"Data Science",
"Machine Learning, Artificial Intelligence",
"Software Engineering"]
```

```python
# creating input dictionary
for i in range(len(mp)):
    for j in range(len(Lo)):
        if Lo[j] in mp.Location[i]:
            P[i,"L",j]=1
        else:
            P[i,"L",j]=0
    for k in range(len(Ti)):
        if Ti[k] in mp.Time[i]:
            P[i,"T",k]=1
        else:
            P[i,"T",k]=0
    for f in range(len(FuncEx)):
        if FuncEx[f] in mp.Func_expertise[i]:
            P[i,"FE",f]=1
        else:
            P[i,"FE",f]=0
        if FuncIn[f] in mp.Func_interest[i]:
            P[i,"FI",f]=1
        else:
            P[i,"FI",f]=0
    for ids in range(len(IndEx)):
        if IndEx[ids] in mp.Ind_expertise[i]:
            P[i,"IE",ids]=1
        else:
            P[i,"IE",ids]=0
        if IndIn[ids] in mp.Ind_interest[i]:
            P[i,"II",ids]=1
        else:
            P[i,"II",ids]=0
    for a in range(len(AreaEx)):
        if AreaEx[a] in mp.Area_expertise[i]:
            P[i,"AE",a]=1
        else:
            P[i,"AE",a]=0
        if AreaIn[a] in mp.Area_interest[i]:
            P[i,"AI",a]=1
        else:
            P[i,"AI",a]=0
                            
    for tee in range(len(ForMentee)):
        if ForMentee[tee] in mp.For_mentee[i]:
            P[i,"Mentee",tee]=1
        else:
            P[i,"Mentee",tee]=0
        if FromMentor[tee] in mp.From_mentor[i]:
            P[i,"Mentor",tee]=1
        else:
            P[i,"Mentor",tee]=0 
```

```python
I=mp.index  
J=mp.index
P
```

```python
GRB=grb.GRB
mod=grb.Model()
mod.setParam("NodefileStart",0.1)
mod.setParam("TimeLimit",2500)
x={}
## decision Variables
for i in I:
    for j in J:
        x[i,j]=mod.addVar(lb=0,vtype=GRB.BINARY,name='x[{0},{1}]'.format(i,j))
## constraints
for i in I:
    for j in J:
        if i==j:
            mod.addConstr(x[i,j]==0)        
        else:
            mod.addConstr(x[i,j]-x[j,i]==0) # A&B and B&A are the same
            mod.addConstr(x[i,j]<=grb.quicksum(P[i,"L",k]*P[j,"L",k] for k in range(len(Lo)))) # location has to be at least one match
            mod.addConstr(x[i,j]<=grb.quicksum(P[i,"T",k]*P[j,"T",k] for k in range(len(Ti)))) # time has to be at least one match
            mod.addConstr(x[i,j]<=grb.quicksum(P[i,"FE",f]*P[j,"FI",f]  + P[i,"IE",ids]*P[j,"II",ids] + P[i,"AE",a]*P[j,"AI",a] + P[i,"Mentee",tee]*P[j,"Mentor",tee]
                               for f in range(len(FuncEx)) for ids in range(len(AreaEx)) for a in range(len(AreaEx))))
                                # expertis, interest, mentor/mentee support area at least a match
    mod.addConstr(grb.quicksum(x[i,j] for j in J)==1)

## set objective : allow maximun number of common locations/time/interests/expertis across all participants
mod.setObjective(grb.quicksum(x[i,j]*P[i,"T",k]*P[j,"T",k] for k in range(len(Ti)) for j in J for i in I)+
                 grb.quicksum(x[i,j]*P[i,"L",k]*P[j,"L",k] for k in range(len(Lo)) for j in J for i in I)+
                 grb.quicksum(x[i,j]*P[i,"FE",k]*P[j,"FI",k] for k in range(len(FuncEx)) for j in J for i in I)+
                 grb.quicksum(x[i,j]*P[i,"IE",k]*P[j,"II",k] for k in range(len(IndEx)) for j in J for i in I)+
                 grb.quicksum(x[i,j]*P[i,"AE",k]*P[j,"AI",k] for k in range(len(AreaEx)) for j in J for i in I)+
                 grb.quicksum(x[i,j]*P[i,"Mentee",k]*P[j,"Mentor",k] for k in range(len(ForMentee)) for j in J for i in I),
                 sense=grb.GRB.MAXIMIZE)
mod.setParam('OutputFlag',True)
mod.optimize()

print('\n-------RESULTS--------------')
print('\n Optimal Value={0}'.format(mod.ObjVal))
print('\n Optimal Solution')

for var in mod.getVars():
    print('\t{0}={1}'.format(var.VarName,var.x))
```

```python
## optimized results
df=pd.DataFrame()
for i in I:
    for j in J:
        df.loc[i,j]=x[i,j].x
df
```
